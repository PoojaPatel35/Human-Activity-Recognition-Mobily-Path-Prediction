```python
def preprocess(path,name,activity):
    dfname = pd.read_csv(path)
    dfname.insert(loc=0, column='UserName', value=name)
    dfname.insert(loc=4, column='Activity', value=activity)
    dfname = dfname.fillna(0)
    dfname.to_csv (path, index = None, header=True) #Don't forget to add '.csv' at the end of the path
    return dfname
```
